```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n_samples = 10
```

```python
u = np.random.uniform(size=n_samples, low=0.0, high=1.0)
```

```python
u
```

```python
M = u.sum() / u.size
M
```

```python
D = (u - M) ** 2
D = D.sum() / D.size
D
```

```python
S = np.sqrt(D)
S
```

```python
def calculate_K(u: np.ndarray, M: np.float64):
    n = u.size
    K = np.zeros(n)
    for f in range(n):
        numerator = np.sum((u[:n-f] - M) * (u[f:] - M))
        denominator = np.sum((u - M) ** 2)
        K[f] = numerator / denominator
    return K
```

```python
K = calculate_K(u, M)
K
```

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
plot_acf(K, color='red', title='K function', ax=ax, fft=True)
plt.show()
```
