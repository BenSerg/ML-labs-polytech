```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/svmdata6.txt', sep='\t')
```

```python
df.head()
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR(kernel='rbf', C=1)
svr.fit(pd.DataFrame(df['X']), df['Y'])
```

```python
plt.scatter(df['X'], df['Y'], color='blue')
plt.plot(df['X'], svr.predict(pd.DataFrame(df['X'])), color='red', linewidth=3)
```
