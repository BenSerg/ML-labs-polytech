```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/eustock.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 8))
len(df)
for i in df.columns:
    plt.plot(range(len(df)), df[i], label=i)
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression
plt.figure(figsize=(10, 10))
for i in df.columns:
    plt.figure()
    model = LinearRegression()
    model.fit(np.array(range(len(df))).reshape(-1, 1), df[i])
    plt.plot(range(len(df)), df[i], label=i)
    
    plt.plot(range(len(df)), model.predict(np.array(range(len(df))).reshape(-1, 1)), label=f'{i} regression', linestyle='--')
    plt.legend()
```
