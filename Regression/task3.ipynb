```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/cygage.txt', sep='\t')
```

```python
df.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X, y, weights = df['calAge'], df['Depth'], df['Weight']
weighted_lin_model = LinearRegression()
lin_model = LinearRegression()
```

```python
weighted_lin_model.fit(pd.DataFrame(X), y, sample_weight=weights)
lin_model.fit(pd.DataFrame(X), y)
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, weighted_lin_model.predict(pd.DataFrame(X)), color='red')
plt.plot(X, lin_model.predict(pd.DataFrame(X)), color='green')
plt.legend(['Data', 'Weighted Linear Regression', 'Default Linear Regression'])
```
