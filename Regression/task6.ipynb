```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/JohnsonJohnson.csv')
```

```python
df.head()
df['year'] = df['index'].apply(lambda x: x.split()[0])
df['quartil'] =df['index'].apply(lambda x: int(x.split()[1][1]))
df = df.drop('index', axis=1)
df.head()
```

```python
df.head()
X = df.drop('value', axis=1)
y = df['value']
X.head()
```

```python
df.groupby('quartil').
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
def quartile_plot(quartile):
    x = X[X['quartil'] == quartile]
    plt.title(f'Q{quartile}', fontweight='bold')
    plt.scatter(df[df['quartil'] == quartile]['year'], df[df["quartil"] == quartile]['value'], color='blue')
    plt.plot(x['year'], model.predict(x), color='black')
    plt.xticks(rotation=90);
```

```python
quartile_plot(1)
```

```python
quartile_plot(2)
```

```python
quartile_plot(3)
```

```python
quartile_plot(4)
```
