```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/reglab1.txt', sep='\t')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X, Y, Z = pd.DataFrame(df['x']), pd.DataFrame(df['y']), pd.DataFrame(df['z'])
```

```python
from sklearn.linear_model import LinearRegression
lin_reg_xy = LinearRegression()
```

```python
lin_reg_xy.fit(X, Y)
```

```python
sns.scatterplot(x=df['x'], y=df['y'], color='blue')
plt.plot(X, lin_reg_xy.predict(X), color='black')
plt.savefig('data/xy.png')
```

```python
lin_reg_xz = LinearRegression()
lin_reg_xz.fit(X, Z)
```

```python
sns.scatterplot(x=df['x'], y=df['z'], color='blue')
plt.plot(X, lin_reg_xz.predict(X), color='black')
plt.savefig('data/xz.png')
```

```python
lin_reg_yz = LinearRegression()
lin_reg_yz.fit(Y, Z)
```

```python
sns.scatterplot(x=df['y'], y=df['z'], color='blue')
plt.plot(Y, lin_reg_yz.predict(Y), color='black')
plt.savefig('data/yz.png')
```
