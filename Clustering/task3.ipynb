```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import os

from sklearn.cluster import KMeans
```

```python
image_as_array = mpimg.imread('/home/serg/PycharmProjects/Ml labs/Clustering/batman.jpg')
```

```python
plt.imshow(image_as_array)
```

```python
h, w, c = image_as_array.shape
```

```python
h, w, c
```

```python
image_as_array_2d = image_as_array.reshape(h * w, c)
```

```python
model = KMeans(n_clusters=5, n_init=10)
```

```python
labels = model.fit_predict(image_as_array_2d)
```

```python
labels
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
new_image = np.reshape(rgb_codes[labels], (h, w, c))
```

```python
plt.imshow(new_image)
```
